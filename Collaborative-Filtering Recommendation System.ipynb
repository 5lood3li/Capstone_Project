```python
#imports
import numpy as np
import random
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import plotly.express as px
import plotly.graph_objects as go


from scipy.sparse.linalg import svds

from sklearn.metrics import silhouette_score
from sklearn.metrics.pairwise import cosine_similarity, linear_kernel
```

```python
df=pd.read_csv("capstone_project_dataset.csv")
```

```python
pd.set_option("display.max_rows", None)
df
```

```python
# Restaurants groubed by their review score
df.groupby('Resturant Name')['Review Score'].count().sort_values(ascending = False)
```

```python
# Creating the Matrix by using the Pivot Table Function
df_rating = df.pivot_table(index = 'User ID', columns = 'Resturant Name', values = 'Review Score')
df_rating
```

```python
# Normalizing the Rating Scores (sice we are using centered cosine similarity)

def mean_center_rows(df1):
    return (df1.T - df1.mean(axis = 1)).T

df_rating = mean_center_rows(df_rating)
df_rating
```

```python
# Filling all Null Values with 0

df_rating = df_rating.fillna(0)
df_rating
```

```python
df_rating.isnull().sum()
```

## Cosine Similarity

```python
# Converting the Matrix DataFrame into a NumPy array

df_matrix = df_rating.to_numpy()
df_matrix[:5]
```

```python
df_matrix.shape
```

# User- Item collaborative filtering

## Truncated SVD

```python
# Applying Singular Value Decomposition (SVD)

#The number of factors to factor the user-item matrix.

NUMBER_OF_FACTORS_MF = 5

#Performs matrix factorization of the original user item matrix

U, sigma, Vt = svds(df_matrix, k = NUMBER_OF_FACTORS_MF)

print(U.shape)
print(Vt.shape)

sigma = np.diag(sigma)
sigma.shape
```

```python
# Overview of user ratings across all Restaurants in Toronto

all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) 
all_user_predicted_ratings
```

```python
# Converting the reconstructed matrix back to a Pandas dataframe
# List of first 10 Customer User_ids in the Matrix
user_ids = list(df_rating.index)

cf_preds_df = pd.DataFrame(all_user_predicted_ratings, columns = df_rating.columns, index=user_ids).transpose()
print(cf_preds_df.shape)
cf_preds_df
```

```python
cf_preds_df.iloc[:,31]
```

# Item- Item collaborative filtering

```python
# Creating Item-Item Matrix based on Cosine Similarity

item_item_matrix = cosine_similarity(cf_preds_df)
item_item_matrix= pd.DataFrame(item_item_matrix, columns=cf_preds_df.index, index = cf_preds_df.index)
item_item_matrix
```

```python
# Creating Collaborative Filtering Function for Restaurant-Restaurant Recommendation System

def cf_recommender(restaurant):
    
    
    restaurant_ratings = cf_preds_df.T[restaurant]
    similar_restaurant_ratings = cf_preds_df.T.corrwith(restaurant_ratings)
    corr_ratings = pd.DataFrame(similar_restaurant_ratings, columns=['Correlation'])
    corr_ratings.dropna(inplace=True)
    
    
    ratings_sim = item_item_matrix[restaurant]
    
   
    ratings_sim = ratings_sim[ratings_sim>0]
    
    """Generate Top 5 Recommended Restaurants"""
    return ratings_sim.sort_values(ascending= False).head(6)[1:]
```

Let's test out our Collaborative Filtering Model on 2 Restaurants: Queen Of India and Shawarma House to see how the system works

```python
# Top 5 Recommended Restaurants that people should visit if they had previously visited Shawarma House

restaurant_recommendations = cf_recommender('Shawarma House')
restaurant_recommendations = pd.DataFrame(data = restaurant_recommendations)
restaurant_recommendations
```
